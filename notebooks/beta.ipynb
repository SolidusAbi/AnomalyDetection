```python
from scipy.stats import beta
from matplotlib import pyplot as plt
x = np.linspace(0, 1.0, 100)
y1 = beta.pdf(x, 2.75, 2.75)
y2 = beta.pdf(x, 3.25, 3.25)
y3 = beta.pdf(x, 3, 3)

plt.plot(x, y1, "-", x, y2, "r--", x, y3, "g--")
plt.show()
```
